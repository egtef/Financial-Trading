```
import bt
#import pandas_datareader as pdr
bt_data=bt.get('goog, amzn, tsla', start='2020-06-01', end='2020-12-01') # directly fetch the closing data
bt_strategy=bt.Strategy('trading_Weekly',
                       [bt.algos.RunWeekly(), #run weekly
                        bt.algos.SelectAll(), # all data
                        bt.algos.WeighEqually(), #add equally amount of capital to each stock
                        bt.algos.Rebalance()
                       ]) # we apply certain strategy than run weekly on portfolio that hold several stocks
bt_test=bt.Backtest(bt_strategy, bt_data)
bt_run=bt.run(bt_test)
```

```
bt_run.plot();
```

- During the last 6 months of 2020 the portfolio is increased from 100 to 180

## Use technical indicators to read patterns in the historical data and anticipate possible future price movments.

1) Trend indicators: measure the direction or strength of a trend.

Benifits: moving out with the price to check, smooth out the data to better indicate the price direction.

* working smoother on longer back period than the shorter period

```
import talib
# Calculate two different moving averages on google dataset
# Calculate 12-day EMA
goog['EMA_12'] = talib.EMA(goog['Close'], timeperiod=12)
# Calculate 26-day EMA
goog['EMA_50'] = talib.EMA(goog['Close'], timeperiod=50)
# Plot the EMAs with price
plt.plot(goog['EMA_12'], label='EMA_12')
plt.plot(goog['EMA_50'], label='EMA_50')
plt.plot(goog['Close'], label='Close')

# Customize and show the plot
plt.legend()
plt.title('EMAs')
plt.show()
```

The shorter EMA of 12 days moves faster with the price than the EMA 50. hence comparing the gap between them can reveal how fast the price is changing.

```
# Calculate EMA and SMA
# Calculate 50-day EMA
goog['EMA'] = talib.EMA(goog['Close'], timeperiod=50)
# Calculate 50-day SMA
goog['SMA'] = talib.SMA(goog['Close'], timeperiod=50)
# Plot the EMAs with price
plt.plot(goog['EMA'], label='EMA_50')
plt.plot(goog['SMA'], label='SMA_50')
plt.plot(goog['Close'], label='Close')

# Customize and show the plot
plt.legend()
plt.title('EMA vs SMA')
plt.show()
```

The EMA 50 is more responsive to the price spikes and less smooth than the SMA. using EMA in shorter trading timeframe would be optimum as it's more sensitive to the recent price fluctiations.

```
# using ADX to measure the strength (Average directional movement index )
```

```
# Calculate the ADX with the default time period on tesla dataframe
tsla['ADX_14'] = talib.ADX(tsla['High'],
                            tsla['Low'], 
                            tsla['Close'])

# Calculate the ADX with the time period set to 21
tsla['ADX_21'] = talib.ADX(tsla['High'],
                            tsla['Low'], 
                            tsla['Close'], timeperiod=21)

# Print the last five rows
print(tsla.tail(5))
```

```
# Calculate ADX, Typically an ADX value above 25 indicates that a trending market is present
tsla['ADX'] = talib.ADX(tsla['High'], tsla['Low'], tsla['Close'])

# Create subplots
fig, (ax1, ax2) = plt.subplots(2)

# plot ADX with the price
ax1.set_ylabel('Price')
ax1.plot(tsla['Close'])
ax2.set_ylabel('ADX')
ax2.plot(tsla['ADX'], color='red')

ax1.set_title('Price and ADX')
plt.show()
```

in Sept 2002 the stock wasn't trending but moving sideways, the ADX indicator started to fall.

```
# Momentum indicator RSI, Relative strength index. 
# Momentum measures the rate of the rising or falling in the asset prices.

# Calculate RSI with the default time period
tsla['RSI_14'] = talib.RSI(tsla['Close'])

# Calculate RSI with a time period of 21
tsla['RSI_21'] = talib.RSI(tsla['Close'], timeperiod=21)

# Print the last five rows
print(tsla.tail(5))
```

```
# Calculate RSI
tsla['RSI'] = talib.RSI(tsla['Close'])

# Create subplots
fig, (ax1, ax2) = plt.subplots(2)
# plot RSI with the price
ax1.set_ylabel('Price')
ax1.plot(tsla['Close'])
ax2.set_ylabel('RSI')
ax2.plot(tsla['RSI'], color='orangered')

ax1.set_title('Price and RSI')
plt.show()
```

In Febrauray 2016 the RSI was below 30 (Oversold), it's more likley to rally afterwards. While in Jan 2020 it was above 70 (Over bought) which reveals probability of moving down afterwards.

```
# Bollinger Bands. volatility indicator
# the wider the bands the more volatile the asset price

# Define the Bollinger Bands with 1-sd
upper_1sd, mid_1sd, lower_1sd = talib.BBANDS(bdata['Close'],
                                             nbdevup=1,
                                             nbdevdn=1,
                                             timeperiod=20)
# Plot the upper and lower Bollinger Bands 
plt.plot(bdata['Close'], color='green', label='Price')
plt.plot(upper_1sd, color='tomato', label="Upper 1sd")
plt.plot(lower_1sd, color='blue', label='Lower 1sd')

# Customize and show the plot
plt.legend(loc='upper left')
plt.title('Bollinger Bands (1sd)')
plt.show()
```

```
# Bollinger Bands. volatility indicator
# the wider the bands the more volatile the asset price

# Define the Bollinger Bands with 2-sd
upper_2sd, mid_2sd, lower_2sd = talib.BBANDS(bdata['Close'],
                                             nbdevup=2,
                                             nbdevdn=2,
                                             timeperiod=20)
# Plot the upper and lower Bollinger Bands 
plt.plot(bdata['Close'], color='green', label='Price')
plt.plot(upper_1sd, color='tomato', label="Upper 2sd")
plt.plot(lower_1sd, color='blue', label='Lower 2sd')

# Customize and show the plot
plt.legend(loc='upper left')
plt.title('Bollinger Bands (2sd)')
plt.show()
```

Yet the Bitcoin was very volatile, and its prices frequently moved out of the 2-sd Bollinger Bands. The plot demonstrated Bitcoin's price peak and crash around Dec 2018, indicating tremendous volatility risk.

#Trading signals: it trigers to buy the asset when the price is greater than the SMA and exit the trade when the price is lower than the SMA.

```
# Construct the signal
appl_data=bt.get('aapl', start='2019-11-01', end='2020-12-01') # directly fetch the closing data
```

```
#sma=talib.SMA(appl_data['Close'], timeperiod=20)
sma=appl_data.rolling(20).mean()
```

```
# Define the strategy
appl_data_strategy = bt.Strategy('AboveSMA', 
                          [bt.algos.SelectWhere(appl_data > sma),
                           bt.algos.WeighEqually(),
                           bt.algos.Rebalance()])
```

```
# Create the backtest and run it
bt_backtest = bt.Backtest(appl_data_strategy, appl_data)
bt_result = bt.run(bt_backtest)
```

```
# Plot the backtest result
bt_result.plot(title='Backtest result')
plt.show()
```

The results was pretty good generated based on backtesting the historical period from Nov 2019 up to Nov 2020 demonstrates a long signal which trigers to buy
